## Aggregating the variables by grouping the countries so that we get an average of variables over the year 2000-2015

```python
# Numerical columns in the df
numerical_cols = df.select_dtypes(include='number').columns
```

```python
# Group the data by country
grouped_df = df.groupby('Country')

# Create an empty DataFrame to store the aggregated results
aggregated_df = pd.DataFrame()

# Add the 'Country' column to the aggregated DataFrame
aggregated_df['Country'] = grouped_df['Country'].first()

# Iterate over the numerical variables and aggregate them
for var in numerical_cols:
    # Calculate the mean for each variable within each country group
    aggregated_var = grouped_df[var].mean()
    
    # Add the aggregated variable to the DataFrame
    aggregated_df[var] = aggregated_var
```

```python
aggregated_df.shape
```

```python
aggregated_df.columns
```

```python
# Getting the status of a country
df.loc[df['Country'] == 'Afghanistan', 'Status'].unique()[0]
```

```python
# Iterate over countries in aggregated_df and assigning the status 
for country in aggregated_df['Country']:
    # Filter the original DataFrame based on the current country
    status = df.loc[df['Country'] == country, 'Status'].unique()[0]
    # Assign the 'Status' value to all rows with the same country in merge_df
    aggregated_df.loc[aggregated_df['Country'] == country, 'Status'] = status
```

```python
# Drop the 'Year' column from the aggregated DataFrame 
aggregated_df = aggregated_df.drop('Year', axis=1)
```

```python
aggregated_df.shape
```

```python
aggregated_df.columns
```

```python
# Convert DataFrame to CSV file
aggregated_df.to_csv('aggregated_data.csv', index=False)
```
